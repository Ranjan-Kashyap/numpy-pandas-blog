# 2. Pandas Series

Have a look at the syntax of the following. You can make a series as a list, a tuple, or directly pass the values as in series3.
Make a note that the data type of these values is "object" that means they are read as strings.
Also, in series1 amd series2 you can only give one value whilst multiple values can be given through series2. 

```python
#Series as a list
series1 = pd.Series(list('45er3'))

#Series as a tuple
series3 = pd.Series(tuple('ert')) 

#Series passed directly
series2 = pd.Series([10,np.nan,'Apple',89])
                     

print("Series as a list: \n",series1) 
print("\nSeries as a tuple: \n",series2) 
print("\nSeries passed directly: \n",series3)
```

# 2.a Series' Indices

```python
#A series made with indices

series4 = pd.Series(['Apple','Mango','Guava','Orange'], index = ['I','II','III','IV'])
print("Series4: \n", series4)

#A series made wihtin a range and indices
series5 = pd.Series(range(5), index = list('abcba'))
print("\nSeries5: \n", series5)
```

## 2.b  Fetching series' values

```python
#To fetch multiple values at time, put the index and get the values: 

print("Series4's index: \n" ,series4[['I','IV']])

print("\nSeries5's index:\n" ,series5[['c','b']])

```

# 3. Pandas DataFrame

DataFrames can be made like that in pandas. Have a look at the way indexing is done in these both the codes. 

```python
#Make a dictionary like this

data2 = {'State': ['Himachal','Punjab','Delhi','Bengal'],
         'Language':['Pahari','Punjabi','Hindi','Bengali'],
         'Population':[46,50,23,83]
        }

#DataFrame is made by passing the dictionary
dataFrame = pd.DataFrame(data2)
dataFrame
```

```python
#A dictionary of dictionaries is made

data3 = {'Language': {'Himachal':'Pahari', 'Punjab':'Punjabi','Bengal':'Bengali'},
         'Folk Dance':{'Himachal':'Nati', 'Punjab':'Bhangra', 'Bengal':'Raibenshe'}}

#Output will be alphabetical
dataFrame2 = pd.DataFrame(data3)
dataFrame2
```

## 3.a Fetching DataFrame values

```python
#Both of these dataframe can be called upon by their column names as:
print(dataFrame['Language'],"\n")

print(dataFrame2['Language'])

#The difference between the outputs is signinficant
```

## 3.b Adding new columns to dataFrames

To add a new column to the "dataFrame" we created at first, follow the syntax below. 

```python
#dataFrame Name ['Column name'] = pd.Series([Values])

dataFrame['Districts'] = pd.Series([12,26,23,25])
dataFrame
```

## 3.c Basic functions on dataFrames

```python
#Sum of strings can't be calculated. So instead it'll print all the string values.
print("Sum of all columns:\n", dataFrame.sum())

#To print only required columns
print("\nSum of only Population column:\t", dataFrame['Population'].sum())
print("\nSum of specific columns:\n", dataFrame[['Population','Districts']].sum())

print("\n\nMean:", dataFrame.mean())
```

```python
#To print all the basic functions

print("Describe Features :", dataFrame.describe())
```

## 4.c Smart File functions

Lambda function is a single line function that can help us do complex operations in one line.
To add a column with certain conditions use lambda function as:

```python
#We are making the "Level" column on the basis of "Profit" column. 
#So, we need to use file.Profit.apply

file["Level"] = file.Profit.apply(lambda x:'Level A' if x>150000 
                                  else('Level B' if x>150000 and x<100000
                                      else ('Level C')))

file.head(10)
```

```python
#To check rows on basis of a conditon
file[file["Level"]=="Level C"]   
```

```python
#To get only specific values on basis of a condition

#We will get the values of the profit column based on the condition that the level should be Level C.
file[file["Level"]=="Level C"]["Profit"]  
                                        
```

```python
#To delete a column, do this:

del file["Country"]
file.head()
```

```python
#To delete multiple columns
#Usually delete will look up for values in rows.
#axis = 1 will guide the program to look in columns.

file.drop(['Level','State'], axis = 1)      
file.head()
```
